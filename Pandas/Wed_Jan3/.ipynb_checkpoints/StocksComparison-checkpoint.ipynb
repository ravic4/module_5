```python
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
#Reading FB 
fb = Path("./fb_data.csv")
fb_df = pd.read_csv(fb, index_col="date", parse_dates=True, infer_datetime_format=True)

#Reading Twitter
twtr = Path("./twtr_data.csv")
twtr_df = pd.read_csv(twtr, index_col="date", parse_dates=True, infer_datetime_format=True)

#Reading Snapchat
snap = Path("./snap_data.csv")
snap_df = pd.read_csv(snap, index_col="date", parse_dates=True, infer_datetime_format=True)

#Reading SPY500 
spy = Path("./sp500_data.csv")
spy_df = pd.read_csv(spy, index_col = "date", parse_dates = True, infer_datetime_format = True) 

# #Testing
# fb_test = fb_df.head()
# twtr_test = twtr_df.head()
# snap_test = snap_df.head()
# spy_test = spy_df.head()

# print(fb_test, twtr_test, snap_test, spy_test) 
```

```python
# Create a new pivot table where the columns are the closing prices for each ticker
combined_df = pd.concat([fb_df, twtr_df, snap_df, spy_df], axis="columns", join="inner")

# Sort datetime index in ascending order (past to present)
combined_df.sort_index(inplace=True)

# Set column names to  and 'S&P 500'
combined_df.columns = ["Facebook","Twitter","Snap", "S&P 500"]

# Display a few rows
combined_df.head()
```

```python
daily_returns = combined_df.pct_change().dropna() #source: https://www.digitalocean.com/community/tutorials/pandas-dropna-drop-null-na-values-from-dataframe 
daily_returns.head() 
```
