## 5. Group DataFrame by `Occupation` and Calculate Average Salary and Age

```python
# Calculate average Salary and Age for each Occupation
# occupation_cleaned = cleaned_people.groupby('Salary')

people_occu = cleaned_people.groupby(['Occupation'])[['Salary','Age']]
people_occu.mean()
```
